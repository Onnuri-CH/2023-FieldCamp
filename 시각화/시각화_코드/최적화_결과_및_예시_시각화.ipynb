```python
from pyproj import Transformer
import pyproj
import numpy as np
import pandas  as pd
from shapely.geometry import Point as point
import geopandas as gpd
import folium
import json
from geopy.distance import geodesic

dong = pd.read_csv('/content/drive/MyDrive/FIELD/location_east.csv')
jung= pd.read_csv('/content/drive/MyDrive/FIELD/location_middle.csv')
```

```python
entertainment_df = pd.concat([dong, jung], ignore_index=True)
entertainment_df
```

```python
# shp 파일 경로
shp_path = '/content/drive/MyDrive/FIELD/LARD_ADM_SECT_SGG_28.shp'

# GeoDataFrame 생성
gdf = gpd.read_file(shp_path)
```

```python
gdf['gid'] = ['Ongjin-gun', 'Ganghwa-gun', 'Seo-gu', 'Gyeyang-gu', 'Bupyeong-gu', 'Namdong-gu', 'Yeonsu-gu', 'Nam-gu', 'Dong-gu', 'Joong-gu','ㅇ']
gdf=gdf.drop(10)
gdf
```

```python
gdf.crs = 'EPSG:5179'
```

```python
# 지도 생성
m = folium.Map(location=[37.5757, 126.6064], zoom_start=15, tiles='cartodbpositron')

folium.Choropleth(
    geo_data=gdf,
    fill_color = '#87CEEB').add_to(m)

for index, row in entertainment_df.iterrows():
    folium.Circle(
        location=[row['위도'], row['경도']],
        radius=200,
        color='Black',
        fill=False,
        fill_opacity=0.3
    ).add_to(m)

m
```

```python
index_to_select = [1,7,11,20,23,27,30,32,34,36,39,43,44,45,47]
```

```python
assign = assign.reset_index(drop=True)
assign
```

```python
lat=assign['Y']
long=assign['X']
```

```python
coords = []
for i in range(len(assign)):
    x = lat[i]
    y = long[i]
    coords.append([x, y])

for i in range(len(coords)):
    folium.Marker(
        location = coords[i],
        popup=assign['학교 id'][i],
        icon=folium.Icon(color='red')
    ).add_to(m)

m
```
