```python
from folium import plugins

data = sdf.drop('수요지 id', axis = 1)
data2 = data.drop('초등학령인구',axis = 1)
data.t=data2.reindex(['Y','X'], axis = 1)
data3 = np.array (data.t)
population = list(sdf['초등학령인구'])

popups = population

m=folium.Map (location=[37.4554, 126.6334],
             zoom_start=12,
             tiles='OpenStreetMap')
plugins.MarkerCluster(data3, popups=population).add_to(m)

# Create a minimap with custom width and height
minimap = plugins.MiniMap(toggle_display=True, width=300, height=300)

# Add the minimap to the map
minimap.add_to(m)

m
```

```python
# Display the map
m.save('map_with_minimap.html')
m
```

```python
from folium.plugins import HeatMap
m = folium.Map(
    location=(37.4554, 126.6334),
    zoom_start=12,
    tiles = 'CartoDB Positron'
)

HeatMap(data3).add_to(m)
m

```

```python
# Save the map to an HTML file
output_file = 'choropleth_map_with_title.html'
m.save(output_file)

print(f'시각화 자료가 {output_file} 파일로 저장되었습니다.')
```
